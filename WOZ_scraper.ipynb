```python
import pandas as pd
import numpy as np
import time
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Firefox()
```

```python
driver.get("https://www.wozwaardeloket.nl/")
```

```python
print(driver.title)
```

```python
#klik op het accepteren van de voorwaarden
driver.find_element_by_css_selector('#accept > fieldset > input.ui-button.ui-widget.ui-state-default.ui-corner-all').click()
```

```python


def findWOZ(address):
    date_element = driver.find_element_by_css_selector('#zoeklocatie')
    date_element.click()
    date_element.clear()
    date_element.send_keys(address)
    
    Zoeken_button = driver.find_element_by_css_selector('#zoeklocatieSubmit')
    Zoeken_button.click()
    time.sleep(4)
    
    WOZ_table = driver.find_element_by_css_selector('#polygoon_wozwaarde_table')
    WOZText= WOZ_table.get_attribute("outerHTML")
    WOZText = BeautifulSoup(WOZText, 'html.parser')
    
    gebruik_table = driver.find_element_by_css_selector('#polygoon_gebruik_table')
    gebruikText = gebruik_table.get_attribute("outerHTML")
    gebruikText = BeautifulSoup(gebruikText, 'html.parser')
    
    WOZText_result = []

    for i,value in enumerate(WOZText.find_all('td')):
        if int(i/2) != i/2:
            WOZText_result.append(value.get_text())
    
    print(address)
    print(WOZText)

    return WOZText_result
```

```python
adressen = pd.read_excel('contacts_cleaned.xlsx')
```

```python
uitwaarden = []

for index, row in adressen.iterrows():
    print (row["Naam"], row["Adres"])
    Adres = row["Adres"]
    try:
        WOZText = findWOZ(Adres)
    except Exception as e:
        print(e)
        driver.close()
        driver = webdriver.Firefox()
        driver.get("https://www.wozwaardeloket.nl/")
        driver.find_element_by_css_selector('#accept > fieldset > input.ui-button.ui-widget.ui-state-default.ui-corner-all').click()
        WOZText = findWOZ(Adres)
        
    result = (row["Naam"], Adres, WOZText)
    print(result)
    uitwaarden.append((row["Naam"], Adres, WOZText))
  
```
